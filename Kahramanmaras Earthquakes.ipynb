```python
#separating data from time column
data['time_new'] = data['time'].str[:10]
data.head()
```

```python
column_name = "mag"
bin_size = 1
bins = np.arange(min(data[column_name]), max(data[column_name]) + bin_size, bin_size)
data['range'] = pd.cut(data[column_name], bins=bins)
grouped_data = data.groupby('range').size()
percentages = (grouped_data / grouped_data.sum()) * 100
plt.figure(figsize=(9, 9))
plt.pie(percentages, labels=percentages.index, autopct='%1.1f%%')
plt.legend(title="Range", labels=[f"{bins[i]:.1f}-{bins[i+1]:.1f}" for i in range(len(bins)-1)], loc="upper right")
plt.title('Distribution of Earthquake Magnitudes')
plt.show()
```

```python
data['time_new'] = pd.to_datetime(data['time_new'])
data_type = data['time_new'].dtype
print("type:", data_type)
```

```python
grouped_data = data.groupby(data['time_new'].dt.to_period('M')).size()
plt.figure(figsize=(8, 8))
plt.pie(grouped_data, labels=grouped_data.index, autopct='%1.1f%%', startangle=140)
plt.axis('equal')
plt.title('Distribution of Earthquakes by Date') 
plt.legend(title='Date', loc='upper right', labels=grouped_data.index.strftime('%Y-%m'))
plt.show()
```

```python
columns_name = 'mag'
ranges = [3,3.5,4, 4.5,5, 5.5,6, 6.5,7, 7.5,8]
veri_sayilari, _ = pd.cut(data[columns_name], bins=ranges, include_lowest=True, right=False, retbins=True)
veri_sayilari.value_counts().sort_index().plot(kind='bar', color='purple')
plt.xlabel('Magnitude')
plt.ylabel('Number of Earthquakes')
plt.title('Number of Earthquakes by Magnitudes')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
